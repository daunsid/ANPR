## Config


```python
from pathlib import Path
DATA_PATH = Path('/kaggle/input/anpr-dataset/anpr-dataset')
DATA_CONFIG = DATA_PATH/"data.yaml"
YOLO_DIR = 'yolo'

YOLO_MODEL_NAME = "yolo11n.pt"

BATCH_SIZE = 100
DEVICE = "cuda"
IMG_SIZE = 640
EPOCHS = 10
```

### Dataset is already in yolo format

```python
print(len(os.listdir(DATA_PATH/"train/images/")))
print(len(os.listdir(DATA_PATH/"test/images/")))
print(len(os.listdir(DATA_PATH/"val/images/")))
```

```python
def train_yolov11(model, data_config, img_size, device, epochs=50, batch_size=50):
    model = YOLO(model)
    model.train(data=data_config, epochs=epochs, batch=batch_size, imgsz=img_size, device=device, workers=4, save=True, save_period=10)
    model.val()
    #print(model)
    print("YOLOv12 training completed on ANPR.")
```

```python
train_yolov11(
    model=YOLO_MODEL_NAME,
    data_config=DATA_CONFIG,
    batch_size=BATCH_SIZE,
    img_size=IMG_SIZE,
    device=DEVICE,
)
```
